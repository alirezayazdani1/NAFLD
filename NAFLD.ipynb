
## Load the lipidome, glycome, fatty acids and hormonal data

```python
pid = pd.read_csv('Data/RawDataPatientIDGroups.csv')
lipids = pd.read_csv('Data/RawDataLipidSpecies.csv')
glycans = pd.read_csv('Data/RawDataGlycans.csv')
fatty = pd.read_csv('Data/RawDataFattyAcids.csv')
hormones = pd.read_csv('Data/RawDataBiochemicalMeasurements.csv')

lipids.fillna(value=0, inplace=True)
glycans.fillna(value=0, inplace=True)
fatty.fillna(value=0, inplace=True)
hormones.fillna(value=0, inplace=True)

# remove the duplicate lipid specie
lip_del = ['SM(d42:5)+H.1']
lipids.drop(lip_del, axis=1, inplace=True)
```

## Utility functions for preprocessing and visualization of data

Parameter 'op' in the 'classify' function follow below encodings:
* 'norm'    : Healthy (all) vs. NAFL vs. NASH
* '-0'      : Healthy (obese) vs. NAFL vs. NASH
* '1v2+3'   : Healthy (all) vs. NAFL and NASH
* '1+2v3'   : Healthy (all) and NAFL vs. NASH
* '2v3'     : NAFL vs. NASH
* 'fibrosis': Healthy (all) vs. Fibrosis

```python
def classify(pid, op):
    if op == 'norm':
        pid['Group'][pid['Group']==0] = 1
        classIds = [1, 2, 3]
    elif op == '-0':
        pid = pid[pid['Group']>0]
        classIds = [1, 2, 3]
    elif op == '1v2+3':
        pid['Group'][pid['Group']==0] = 1
        pid['Group'][pid['Group']==3] = 2
        classIds = [1, 2]
    elif op == '1+2v3':
        pid['Group'][pid['Group']==0] = 1
        pid['Group'][pid['Group']==2] = 1
        pid['Group'][pid['Group']==3] = 2
        classIds = [1, 2]
    elif op == '2v3':
        pid = pid[pid['Group']>1]
        pid['Group'][pid['Group']==2] = 1
        pid['Group'][pid['Group']==3] = 2
        classIds = [1, 2]
    elif op == 'fibrosis':
        A = [2, 3, 4, 6, 8, 12, 14, 18, 19, 21, 23, 24, 26, 30, 32, 42, 45, 58, 60, 64, 68]
        B = [16, 22, 25, 27, 29, 31, 41, 48, 49, 65]
        pid = pd.DataFrame({'Patient ID': A+B,
                            'Group': len(A)*[1]+len(B)*[2]})
        classIds = [1, 2]
    return pid, classIds

# centered log-ratio function
def clr(X):
    x = np.log(X)
    return x - np.mean(x, axis=0)

def prep_data(data, normalized=False):
        # remove variables with more than 1/3 missing data
        for col in data.columns:
            coldata = data[col].to_numpy()
            if len(coldata[np.where(coldata==0)]) >= np.floor(len(coldata)/3):
                data.drop(col, axis=1, inplace=True)
                print('Deleted variable %s' % col)
        # apply clr if data is normalized
        if normalized:
                data = clr(data + data[data>0].min().min()/5)
        # fill missing values with the median for each variable
        else:
            for col in data.columns:
                colmed = data[col][data[col]>0].median()
                data[col] = data[col].where(data[col]!=0, colmed)
        return data

# scale data to have zero mean and unit variance
def standardization(X):
    return preprocessing.scale(X)

# function to plot the PCA (2 components) and 2D tSNE visualization of data
def visualization(X, y, classIds, pca=True, tsne=True):
    ax1 = []; ax2 = []
    fig = plt.figure(figsize=(16, 6))
    n_classes = len(classIds)
    if  n_classes == 3:
        colors = cycle(['red', 'green', 'blue'])
    else:
        colors = cycle(['red', 'green'])
    if pca:
        ax1 = fig.add_subplot(1, 2, 1)
        model = PCA(n_components=2)
        X_pca = model.fit_transform(X)
        for i, color in zip(range(1, n_classes+1), colors):
            ax1.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label='Group {}'.format(i), color=color)
        ax1.set_title('PCA Visualization of Latent Components', fontsize=16)
        ax1.set_xlabel('PCA $1^{st}$ Component', fontsize=14)
        ax1.set_ylabel('PCA $2^{nd}$ Component', fontsize=14)
        ax1.legend(loc='best', fontsize=12)
    
    if tsne:
        ax2 = fig.add_subplot(1, 2, 2)
        model = TSNE(n_components=2, init='random', early_exaggeration = 1, 
                     random_state=0, perplexity=50)
        X_tsne = model.fit_transform(X)
        for i, color in zip(range(1, n_classes+1), colors):
            ax2.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label='Group {}'.format(i), color=color)
        ax2.set_title('t-SNE Visualization of Latent Variales', fontsize=16)
        ax2.set_xlabel('t-SNE $1^{st}$ Component', fontsize=14)
        ax2.set_ylabel('t-SNE $2^{nd}$ Component', fontsize=14)
        ax2.legend(loc='best', fontsize=12)
    return fig, [ax1, ax2]
```

## Dimensionality reduction and variable selection

1. Select variables based on k highest scores from ANOVA F-value
2. Recursive Feature Elimination (RFE) based on a linear SVM

```python
def feature_selection(X, y, k, feature_names, op, omics):
    if op == 'FTest':
        sel = SelectKBest(f_classif, k=k)
        X_new = sel.fit_transform(X, y)
        feature_ids = np.argsort(-sel.scores_)
        kk = k

    if op == 'RFE':
        estimator = SVC(C=0.125, kernel='linear', decision_function_shape='ovr')
        sel = RFE(estimator, n_features_to_select=k, step=1)
        X_new = sel.fit_transform(X, y)
        kk = sel.n_features_
        feature_ids = [np.argwhere(sel.support_)[i][0] for i in range(kk)]

    if op == 'RFECV':
        estimator = SVC(C=0.125, kernel='linear', decision_function_shape='ovr')
#        estimator = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2,
#                                           max_features=None, bootstrap=True)
        sel = RFECV(estimator, step=1, min_features_to_select=k, 
                    cv=StratifiedKFold(n_splits=3, random_state=None), 
                    n_jobs=-1)
        X_new = sel.fit_transform(X, y)
        kk = sel.n_features_
        feature_ids = [np.argwhere(sel.support_)[i][0] for i in range(kk)]
        fig = plt.figure(figsize=(7, 6))
        plt.xlabel("Number of features selected")
        plt.ylabel("Cross validation score\n(fraction of correct classifications)")
        plt.plot(range(1, len(sel.grid_scores_)+1), sel.grid_scores_)
        plt.show()
        extent = fig.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
        fig.savefig(fname='Results/{}RFECV.png'.format(omics), dpi=400, bbox_inches=extent.expanded(1.0, 1.0))

    selected = [feature_names[feature_ids[i]] for i in range(kk)]
    print('%d variables from %s selected using %s method:\n' % (kk, omics, op))
    print(selected)
    print()
    return X_new, selected
```

## Fit classifiers (SVM, kNN, random forest) and use cross-validation for their performance evaluation

Funtion 'grid_search' is used for hyperparameter tuning, which is especially important for improving nonlinear SVM classifier accuracy. The reduced lipidomics dataset is used to fine tune the classifier, which is subsequently applied to all omics datasets.

```python
def linearSVM(X=None, Y=None):
    '''C: [10^-5, 10^5]
    '''
    clf = LinearSVC(dual=False, C=0.5, class_weight='balanced')
    if X is None:
        return clf
    clf.fit(X, Y)
    return clf

def kNN(X=None, Y=None):
    '''n_neighbors
    weights: uniform, distance
    '''
    clf = neighbors.KNeighborsClassifier(n_neighbors=6, weights='uniform')
    return clf

def SVM(X=None, Y=None):
    '''C: [10^-5, 10^5]
    '''
    clf = SVC(C=512.0, gamma=0.00048828125, class_weight='balanced')
    if X is None:
        return clf
    clf.fit(X, Y)
    return clf

def random_forest(X=None, Y=None):
    clf = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=12,
                                 class_weight='balanced', max_features=2, bootstrap=True)
    if X is None:
        return clf
    clf.fit(X, Y)
    return clf

def grid_search(X, y, param, method):
    if method == 'forest':
        clf = GridSearchCV(RandomForestClassifier(class_weight='balanced', max_depth=None),
                           param_grid=param, iid=False, cv=5, n_jobs=-1,
                           scoring=make_scorer(balanced_accuracy_score))
        clf.fit(X, y)
    elif method == 'knn':
        clf = GridSearchCV(neighbors.KNeighborsClassifier(weights='uniform'),
                           param_grid=param, iid=False, cv=5, n_jobs=-1,
                           scoring=make_scorer(balanced_accuracy_score))
        clf.fit(X, y)
    elif method == 'linearsvm':
        clf = GridSearchCV(SVC(kernel='linear', class_weight='balanced', decision_function_shape='ovr'),
                           param_grid=param, iid=False, cv=5, n_jobs=-1,
                           scoring=make_scorer(balanced_accuracy_score))
        clf.fit(X, y)

    elif method == 'svm':
        clf = GridSearchCV(SVC(kernel='rbf', class_weight='balanced', decision_function_shape='ovr'),
                           param_grid=param, iid=False, cv=5, n_jobs=-1,
                           scoring=make_scorer(balanced_accuracy_score))
        clf.fit(X, y)

    print("Best estimator for the algorithm found by grid search:")
    print(clf.cv_results_['params'][clf.best_index_])
    
# rbf SVM
# param_grid = {'C': [2**i for i in range(-5, 16)],
#              'gamma': [2**i for i in range(-15, 3)]}

# kNN
# param_grid = {'n_neighbors': range(1, 10, 1)}

# random forest
# param_grid = {'n_estimators': range(10, 200, 10),
#               'max_features': range(2, np.shape(Xlip)[1], 4),
#               'min_samples_split': range(2, np.shape(Xlip)[0], 5)}

# grid_search(Xlip, ylip, param_grid, 'svm')
```

## Single 'omics classification performance based on 3-fold cross-validation repeated 100 times

```python
scoring = {'Accuracy': make_scorer(balanced_accuracy_score, greater_is_better=True),
           'Recall': make_scorer(recall_score, greater_is_better=True, average='weighted'),
           'Spec': make_scorer(specificity_score, greater_is_better=True, average='weighted')}

def model_selection(X, y):
    clfs = {
        'Linear SVM': linearSVM(),
        'kNN': kNN(),
        'SVM': SVM(),
        'Random forest': random_forest(),
        }
    for name, clf in clfs.items():
        scores = cross_validate(clf, X, y,
                                cv=RepeatedStratifiedKFold(n_splits=3, n_repeats=100, random_state=None),
                                scoring=scoring, n_jobs=-1)
        df = pd.DataFrame(scores)
        print('Classifier %s: \t Accuracy \t Sensitivity \t Specificity' % name)
        print('\t \t %0.2f (+/- %0.2f) \t %0.2f (+/- %0.2f) \t %0.2f (+/- %0.2f)' % 
              (df.test_Accuracy.mean(), df.test_Accuracy.std(), df.test_Recall.mean(), df.test_Recall.std(),
               df.test_Spec.mean(), df.test_Spec.std()))
```

```python
print('NAFLD Classification Based on Lipids Data:')
model_selection(Xlip, ylip)

print(50*'-')

print('NAFLD Classification Based on Hormonal Data:')
model_selection(Xhorm, yhorm)

print(50*'-')

print('NAFLD Classification Based on Glycans Data:')
model_selection(Xglyc, yglyc)

print(50*'-')

print('NAFLD Classification Based on Fatty Acids Data:')
model_selection(Xfatty, yfatty)
```

## ROC curves and AUC for single 'omics data
1. Nonlinear SVM with RBF kernel as classifier
2. 3-fold cross-validation repeated 100 times

```python
def ROC(X, y, classIds, axis, title):
    # classification and ROC analysis
    # run classifier with cross-validation
    cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=100, random_state=None)

    # learn to predict each class against the other
    classifier = OneVsRestClassifier(SVM())

    tprs = dict()
    roc_auc = dict()
    mean_fpr = np.linspace(0, 1, 100)
    mean_auc = []
    n_classes = len(classIds)
    if n_classes == 2:
        n_classes = 1
    for i in range(n_classes):
        tprs[i] = []
        roc_auc[i] = []
    
    for train, test in cv.split(X, y):
        y_score = classifier.fit(X[train], y[train]).decision_function(X[test])
        
        # binarize the output
        y_test_label = preprocessing.label_binarize(y[test], classes=classIds)
    
        mean_auc.append(roc_auc_score(y_test_label, y_score, average='weighted'))
    
        # compute ROC curve and area under the curve
        for i in range(n_classes):
            if n_classes == 1:
                fpr, tpr, _ = roc_curve(y_test_label, y_score)
            else:
                fpr, tpr, _ = roc_curve(y_test_label[:, i], y_score[:, i])
            tprs[i].append(interp(mean_fpr, fpr, tpr))
            roc_auc[i].append(auc(fpr, tpr))
    
    # plot all ROC curves
    lw = 2
    colors = cycle(['red', 'green', 'blue'])
    if n_classes == 3:
        for i, color in zip(range(n_classes), colors):
            axis.plot(mean_fpr, np.mean(tprs[i], axis=0), color=color, lw=lw,
                      label=r'%d vs. Other(s) (AUC = %0.2f $\pm$ %0.2f)' 
                      % (classIds[i], np.mean(roc_auc[i]), np.std(roc_auc[i])))
    else:
        for i, color in zip(range(n_classes), colors):
            axis.plot(mean_fpr, np.mean(tprs[i], axis=0), color=color, lw=lw,
                      label=r'%d vs. 2 (AUC = %0.2f $\pm$ %0.2f)' 
                      % (classIds[i], np.mean(roc_auc[i]), np.std(roc_auc[i])))
        
    axis.plot([0, 1], [0, 1], 'k--', lw=lw)
    axis.set_xlim([0.0, 1.0])
    axis.set_ylim([0.0, 1.05])
    axis.set_xlabel('False Positive Rate', fontsize=14)
    axis.set_ylabel('True Positive Rate', fontsize=14)
    axis.set_title(title, fontsize=16)
    axis.legend(loc="lower right", fontsize=12)

    print('%s - AUC: \t %.2f (+/- %0.2f)' % (title, np.mean(mean_auc), np.std(mean_auc)))
```

## Multi-omics analysis using concatenated lipids, hormone and glycans data
